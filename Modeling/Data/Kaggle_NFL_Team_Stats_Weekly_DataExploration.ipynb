```python
# will want df to match with other df for future joins
kaggle_weekly_stats_df['week'] = kaggle_weekly_stats_df['week'].replace('Division', 'Divisional')
kaggle_weekly_stats_df['week'] = kaggle_weekly_stats_df['week'].replace('Superbowl', 'Super Bowl')
kaggle_weekly_stats_df['week'] = kaggle_weekly_stats_df['week'].replace('Wildcard', 'Wild Card')
kaggle_weekly_stats_df['week'] = kaggle_weekly_stats_df['week'].replace('Conference', 'Championship')
kaggle_weekly_stats_df['week'].unique()
```
