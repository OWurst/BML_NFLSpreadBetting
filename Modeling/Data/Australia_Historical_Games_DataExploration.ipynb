```python
import pandas as pd
```

```python
australia_historical_game_outcomes_df = pd.read_excel('Australia_Historical_Game_Outcomes.xlsx', engine='openpyxl')
```

```python
australia_historical_game_outcomes_df.head(5)
```

```python
australia_historical_game_outcomes_df.columns
```

```python
australia_historical_game_outcomes_df = australia_historical_game_outcomes_df[['Date','Home Team', 'Away Team', 'Home Score', 'Away Score', 'Home Line Close']]
```

```python
australia_historical_game_outcomes_df['Score Diff'] = australia_historical_game_outcomes_df['Home Score'] - australia_historical_game_outcomes_df['Away Score']
```

```python
australia_historical_game_outcomes_df.head(5)
```

```python
australia_historical_game_outcomes_df['Home Spread Diff'] = australia_historical_game_outcomes_df['Home Line Close'] + australia_historical_game_outcomes_df['Score Diff']
```

```python
australia_historical_game_outcomes_df.head(25)
```

```python
# home beat spread if home spread diff is greater than 0
# home push spread if home spread diff is equal to 0
# home lose spread if home spread diff is less than 0
australia_historical_game_outcomes_df['Home vs Spread'] = australia_historical_game_outcomes_df['Home Spread Diff'].apply(lambda x: 'Win' if x > 0 else ('Push' if x == 0 else 'Lose'))
```

```python
# get the number of games that home team beat the spread
australia_historical_game_outcomes_df['Home vs Spread'].value_counts()
```

```python
australia_historical_game_outcomes_df['Date'] = pd.to_datetime(australia_historical_game_outcomes_df['Date'])

# season year is the year that the season starts in, games in a season can span multiple years so anythign before march is considered the previous year
australia_historical_game_outcomes_df['Season'] = australia_historical_game_outcomes_df['Date'].apply(lambda x: x.year if x.month >= 3 else x.year - 1)
```

```python
australia_historical_game_outcomes_df[australia_historical_game_outcomes_df['Season'] == 2023].head(20)
```

```python
# Determine the start date for each season
season_start_dates = australia_historical_game_outcomes_df.groupby('Season')['Date'].min()

# Function to calculate the week number
def get_week_number(row):
    season_start_date = season_start_dates[row['Season']]
    delta = row['Date'] - season_start_date
    week_number = (delta.days // 7) + 1
    if week_number > 18:
        return "Playoffs"
    return week_number

# Calculate the week number for each game
australia_historical_game_outcomes_df['Week'] = australia_historical_game_outcomes_df.apply(get_week_number, axis=1)

# Display the updated DataFrame
australia_historical_game_outcomes_df.head()
```

```python
# Assuming 'HomeTeam' and 'AwayTeam' are the column names for the teams in your dataset
# Melt the DataFrame to have a single column for teams
melted_df = australia_historical_game_outcomes_df.melt(id_vars=['Season', 'Week', 'Date'], value_vars=['Home Team', 'Away Team'], var_name='HomeAway', value_name='Team')

# Group by Season, Week, and Team, then count the occurrences
team_week_counts = melted_df.groupby(['Season', 'Week', 'Team']).size().reset_index(name='GameCount')

# Filter to find teams that play more than once in the same week
duplicate_games = team_week_counts[team_week_counts['GameCount'] > 1]

# Display the teams that play more than once in the same week
duplicate_games
```

```python
# remove date column
australia_historical_game_outcomes_df = australia_historical_game_outcomes_df.drop(columns='Date')
```
